```python
# problem statement
''' A telecommunication company wants to reduce customer chern by identifying customer at risk of leaving .
They have historical data on customer behaviour and wants to build to predict which customers are most likely to churn.'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC # svm= support vector machine , SVC = support vector classifier.
from sklearn.metrics import accuracy_score,classification_report # to evaluate the performence of any model
```

```python
# Step 2
data = {'Age':[30,25,35,20,40,55,32,28],'Monthely_charges':[50,60,80,40,100,120,70,55],'Churn':[ 0,1,0,1,0,1,0,1]}
df=pd.DataFrame(data)
df
```

```python
X = df[['Age','Monthely_charges']]
Y= df['Churn']
```

```python
# Step 3
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,random_state=42,test_size=.2)
```

```python
# step 4
svc_model = SVC(kernel='linear',C=1.0)  # default regularization 
svc_model.fit(X_train,Y_train)
```

```python
# Step 5
prediction = svc_model.predict(X_test)
```

```python
accu= accuracy_score(Y_test,prediction)
```

```python
accu
```

```python
# Step 6
report = classification_report(Y_test,prediction)
print(report)
```

```python
# Step 7
user_age=float(input("Enter the custumer age: "))
user_monthly_charges= float(input("Enter you monthely recharches: "))
user_input = np.array([[user_age,user_monthly_charges]])
Prediction = svc_model.predict(user_input)
if Prediction[0] == 0:
    print("The customer is likely  to stay :")
else:
    print("The customer is not likely to churing")
```
