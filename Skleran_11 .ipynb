```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler  # use for find diffrence btw mean and standard deviation 

```

```python
scaler = StandardScaler() # for standardization
scaled_data= scaler.fit_transform(datanew)
```

```python
clustering = AgglomerativeClustering(n_clusters=3,metric ='euclidean',linkage ='ward')
datanew['Cluster'] = clustering.fit_predict(scaled_data)
```

```python
## visulization
plt.scatter(datanew['Annual Income (k$)'],datanew['Spending Score (1-100)'],c=datanew['Cluster'],cmap='viridis')
plt.xlabel(" Annual income")
plt.ylabel(" Spending score")
plt.title(" Graph")
plt.show()
```
