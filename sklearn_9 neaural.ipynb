```python
##  problem 
'''you are tasked with creating a nearal network model to predict whether a student will pass or fail an exam based on two features:
hours studied and previous exam score . the dataset should contains information on hours studied and previous exam scores for a group of students ,
along with their exam outcomes(pass or fail)'''
```

```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score
```

```python
# # for practice 
# data = {'name': ['jay','praveen','abhay','arun'],'hour_studied':[3,2,5,4],'previous_exam_score':[20,72,80,18],'pass_or_fail':[0,1,1,0]}
# data = pd.DataFrame(data)
# data
```

```python
## Step 1
# Student_Name = ['jay','praveen','abhay','arun','vishal','himanshu','aditya','ravi']
hours_studied = [2,4,6,4,1,8,10,12]
prev_exam_score =[ 23,35,50,45,10,68,75,83]
exam_outcome = ['fail','pass','pass','pass','fail','pass','pass','pass']
```

```python
# Step 2

label_encoder = LabelEncoder()
encoded_exam_outcome = label_encoder.fit_transform(exam_outcome)

# student_name = label_encoder.fit_transform(Student_Name)
```

```python
# Step 4
X = np.column_stack((hours_studied,prev_exam_score))
y =  encoded_exam_outcome
```

```python
# Step 5
clf = MLPClassifier(hidden_layer_sizes =(4,),activation='logistic',max_iter =1000,random_state=42)
clf.fit(X,y)
```

```python
## Step 6
new_student_data = np.array([[1,10]]) # hours and pre exam score 

# Step 7
predicted_outcome= clf.predict(new_student_data)
```

```python
# Step 8
predicted_outcome_decode = label_encoder.inverse_transform(predicted_outcome)
```

```python
# step 9
print("Predicted exam outcome for new student {}".format(predicted_outcome_decode))
```
