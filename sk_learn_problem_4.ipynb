```python
''' A retail company wants to predict custumer purchasing behavior based on their age ,salery and past purchase history .
The company aim to use K-nearest neighbors(KNN) alorithm to classify customers into potential model will help the company
understand and target specific customer segments more effectively,thereby increasing sales and customers satisfactions.'''
```

```python
# step 1
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# step 2
data = np.array([[25,50000,2],[30,80000,1],[35,60000,3],[20,30000,2],[40,90000,1],[45,75000,2]])
labels =np.array([1,2,1,0,2,1]) # 0 = low, 1 = medium ,2 = high
data
```

```python
# Step 3
X_train,X_test,Y_train,Y_test= train_test_split(data,labels,test_size=.2,random_state=42)# why use X_train and X_test
```

```python
# Step 4
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
```

```python
# Step 4
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
```

```python
accuracy = knn.score(X_test,Y_test)
```

```python
    print(f"Model accuracy :{accuracy}")
```
