```python
# problem
'''A mall wants to understand its custumer base better to improve its marketing strategy.The goal is to cluster customers into groups 
based on their annual income and spending score,so the marketing team can tailor their campaigns to each cluster's preference and behaviors.'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
# # step 1
# data = pd.read_csv('C:/Users/vsran/Downloads/Mall_Customers.csv')
# data = pd.DataFrame(data)
# data.head(10)

# OR 

# np.random.seed(42)
data ={'AnnualIncome':np.random.randint(20000,100000,100),'spendingScore':np.random.randint(1,100,100)}
data=pd.DataFrame(data)
data
```

```python
# data=pd.get_dummies(data,columns=['Gender'])
plt.scatter(data['AnnualIncome'],data['AnnualIncome'])
plt.title("Customer data Annual Income vs spending score")
plt.xlabel('Annual Income')
plt.ylabel("spending_score")
plt.show()
```

```python
X= data.values
kmeans=KMeans(n_clusters=3,random_state=42)
data['cluster']=kmeans.fit_predict(X)
data.head(10)
```

```python
# data=pd.get_dummies(data,columns=['Gender'])
plt.scatter(data['AnnualIncome'],data['spendingScore'],c =data['cluster'],cmap = 'rainbow')

plt.title("Kmeans clustiring annual income and spending")
plt.xlabel('Annual Income')
plt.ylabel("spending_score")
plt.show()
```

```python
# userINPUt

user_input = {'AnnualIncome':[50000],'spendingScore':[50]}
user_df =pd.DataFrame(user_input)
```

```python
user_cluster = kmeans.predict(user_df)
```

```python
print(f"The user belong to cluster{user_cluster}")
```
