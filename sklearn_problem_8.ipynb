```python
# problem
'''Create a predictive model using gradiant boosting to forcast housing prices based on varius features such as square footage ,number of bedrooms,
number of bathrooms ,and location'''
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# 1 .from sklearn.experimental import enable_hist_gradient_boosting
# 2 .from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Step 1
data = { 'sq_Foot':[1500,2000,1200,1800,1350],
        'NoOfBedRooms': [2,4,3,1,2],
       'NoOfBathRooms':[4,6,5,3,2],
       'Location':['bati','chhatikara','raal','vrindavan','mathura'],
       'Price':[250000,300000,180000,200000,220000]}
```

```python
# step 2
df= pd.DataFrame(data)
df
```

```python
## Step 3
```

```python
X = df.drop('Price',axis=1) # first drop from x
y= df['Price'] # and in y 
```

```python
# Step 4
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state =42,test_size=.2)
model= GradientBoostingRegressor()
model.fit(X_train,y_train)
```

```python
# Step 5
y_pred = model.predict(X_test)
y_pred
```

```python
mse =  mean_squared_error(y_test,y_pred)
print(mse)
```

```python
# Step 6

# user input
print("Enter the details for house prediction : ")
square_Foot = float(input("Square Footage : "))
BedRooms = int(input("Number of bedrooms : "))
BathRooms = int(input("Number of bathrooms : "))
location = input("Location (bati / mathura /vrindavan / raal / chhatikara  :")


# Step 7
input_location = [0,0,0,0,0]

    
if Location == 'chhatikara':  
    input_location [1]= 1
    
elif Location == 'raal':
    input_location [2]=1
    
elif Location == 'vrindavan'  :
    input_location[3] =1
    
elif Location == 'mathura' :
    input_location [4]=1
elif Location == 'bati':
    input_location [0] ==1    


# Step 8
user_input=pd.DataFrame({ 'sq_Foot':[square_Foot],
        'NoOfBedRooms': [BedRooms],
       'NoOfBathRooms':[BathRooms],
                         
       'Location_bati':[input_location[1]],
       'Location_chhatikara':[input_location[2]],
       'Location_raal':[input_location[3]],
       'Location_vrindavan':[input_location[4]],
       'Location_mathura':[input_location[0]]} )



# step 9


predicted_price= model.predict(user_input)
print(f"Predicted Price for the House :{predicted_price[0]}")
```
